# **What** **is** **Pandas**
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

# **Pandas** **Series**
A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

# **Series** **from** **lists**

```python
#string
country = ['India','America','Russia','Australia']
pd.Series(country)
```

```python
#integer
runs = [23,34,60,80]
runs_series = pd.Series(runs)
runs_series
```

```python
#custom index
marks = [45,60,80,90,68]
subjects = ['Analysis','Numerical','C++','Python','DSA']
pd.Series(marks,index=subjects)
```

```python
#name
marks_series = pd.Series(marks,index=subjects,name='Ziyan Marks')
marks_series
```

# **Series** **from** **dictionary**

```python
marks = {
    'Analysis':68,
    'Python':90,
    'C++':80
}
```

```python
marks_series= pd.Series(marks,name='Ziyan Marks')
marks_series
```
